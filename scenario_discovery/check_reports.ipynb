```python
summary_df = pd.read_csv("scenario_discovery_outputs/country_run_summary.csv")
summary_df.head()
```

```python
summary_df = summary_df[
    [
        "country",
        "status",
        "vulnerability_threshold_vs_2022",
        "vulnerability_threshold_vs_ndc_unconditional",
        "selected_density_below_min"
    ]
]
```

```python
summary_df.head()
```

```python
def compute_percentages(df):
    results = {}

    for col in df.columns:
        if col == "country":
            continue
        
        # Get percentage distribution
        pct = (
            df[col]
            .value_counts(normalize=True)
            .mul(100)
            .round(2)
        )
        
        results[col] = pct.to_dict()
    
    return results
```

```python
percentages = compute_percentages(summary_df)
print(percentages)
```
