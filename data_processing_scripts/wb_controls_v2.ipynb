```python
import wbdata
import pandas as pd
import datetime
import os
```

```python
SCRIPT_DIR_PATH = os.getcwd()
ROOT_DIR_PATH = os.path.dirname(SCRIPT_DIR_PATH)
DATA_DIR_PATH = os.path.join(ROOT_DIR_PATH, "data")
PROCESSED_DATA_DIR_PATH = os.path.join(DATA_DIR_PATH, "processed_data")
```

```python
# 1. Define the indicators you want to fetch
indicators = {
    'NY.GDP.PCAP.KD':       'gdp_per_capita',
    'SP.POP.TOTL':          'population',
    'SP.POP.GROW':          'population_growth',
    'EG.USE.PCAP.KG.OE':    'energy_per_capita',
    'EG.FEC.RNEW.ZS':       'renewable_share',
    'EG.GDP.PUSE.KO.PP.KD': 'energy_intensity',
    'SP.URB.TOTL.IN.ZS':    'urbanization_rate',
    'NV.IND.TOTL.ZS':       'industry_size' 
}

# 2. Set your date range using datetime.datetime
date_range = (
    datetime.datetime(1990, 1, 1),
    datetime.datetime(2024, 1, 1)
)

# 3. Fetch the data (using the correct keywords for this version)
df = wbdata.get_dataframe(
    indicators,
    country="all",
    date=date_range,        # was data_date before
    freq='Y',               # annual data; change to 'M' or 'Q' if you need monthly/quarterly
    parse_dates=True,       # was convert_date before
    keep_levels=False,      # drop singleton index levels
    skip_cache=False        # set True to bypass wbdata’s cache
)

# 4. Reset index so “country” and “date” become columns, then rename them
df = df.reset_index().rename(columns={'country': 'iso3', 'date': 'year'})

# 5. Convert the year column format to only the year
df['year'] = df['year'].dt.year
```

```python
df.to_csv(os.path.join(PROCESSED_DATA_DIR_PATH, 'wb_control_vars.csv'), index=False)
```
