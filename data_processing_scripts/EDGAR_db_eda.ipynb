```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
curr_dir_path = os.getcwd()
main_dir_path = os.path.abspath(os.path.join(curr_dir_path, os.pardir))
data_dir_path = os.path.join(main_dir_path, 'data')
raw_data_dir_path = os.path.join(data_dir_path, 'raw_data')
processed_data_dir_path = os.path.join(data_dir_path, 'processed_data')
cw_dir_path = os.path.join(data_dir_path, 'cw')
emissions_data_dir_path = os.path.join(raw_data_dir_path, 'emissions_data')
iea_data_dir_path = os.path.join(raw_data_dir_path, 'IEA_climate_policy_data')
```

```python
emissions_db = pd.read_csv(os.path.join(emissions_data_dir_path, 'CSC-GHG_emissions-April2024_to_calibrate.csv'), encoding='latin1')
emissions_db.head()
```

```python
emissions_db.info()
```

## Create a DB with total emissions data per country and year

```python
total_em_db = emissions_db.drop(columns=['Country',
                                         'EDGAR Country Code',
                                         'Lending category',
                                         'Region',
                                         'CSC Sector',
                                         'CSC Subsector',
                                         'Gas',
                                         'Units'])
total_em_db.head()
```

```python
total_em_db = total_em_db.groupby(['Code', 'Income group']).sum().reset_index()
```

```python
total_em_db.head()
```

```python
# Change total_em_db to wide format
total_em_db_long = pd.melt(total_em_db, id_vars=['Code', 'Income group'], var_name='Year', value_name='Total Emissions')
total_em_db_long.sort_values(by=['Code', 'Year'], inplace=True)
total_em_db_long.reset_index(drop=True, inplace=True)
```

```python
# make cols lowercase and replace spaces with underscores
total_em_db_long.columns = total_em_db_long.columns.str.lower().str.replace(' ', '_')

# rename 'code' to 'iso_alpha_3'
total_em_db_long.rename(columns={'code': 'iso_alpha_3'}, inplace=True)

total_em_db_long
```

```python
total_em_db_long.to_csv(os.path.join(processed_data_dir_path, 'total_emissions_db.csv'), index=False)
```
