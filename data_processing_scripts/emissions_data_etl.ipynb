```python
import pandas as pd
import os
```

```python
CURR_DIR_PATH = os.getcwd()
MAIN_DIR_PATH = os.path.abspath(os.path.join(CURR_DIR_PATH, os.pardir))
DATA_DIR_PATH = os.path.join(MAIN_DIR_PATH, 'data')
RAW_DATA_DIR_PATH = os.path.join(DATA_DIR_PATH, 'raw_data')
PROCESSED_DATA_DIR_PATH = os.path.join(DATA_DIR_PATH, 'processed_data')
EMISSIONS_DATA_DIR_PATH = os.path.join(RAW_DATA_DIR_PATH, 'emissions_data')
```

## 1. Load EDGAR's Emission Data

```python
emissions_df= pd.read_csv(os.path.join(EMISSIONS_DATA_DIR_PATH, 'EDGAR_Emission_Totals.csv'))
emissions_df.head()
```

```python
emissions_df.info()
```

## 2. Create a CSV with total emissions data per country and year (long format)

```python
emissions_df.columns
```

```python
# Drop unnecessary columns
total_em_df = emissions_df.drop(columns=['IPCC_annex', 'C_group_IM24_sh', 'Name', 'Substance'])
total_em_df.head()
```

```python
# Change total_em_db to wide format
total_em_df_long = pd.melt(total_em_df, id_vars=['Country_code_A3'], var_name='Year', value_name='Total Emissions')
total_em_df_long.sort_values(by=['Country_code_A3', 'Year'], inplace=True)
total_em_df_long.reset_index(drop=True, inplace=True)
total_em_df_long.head()
```

```python
# make cols lowercase and replace spaces with underscores
total_em_df_long.columns = total_em_df_long.columns.str.lower().str.replace(' ', '_')

# rename 'country_code_a3' to 'iso_alpha_3'
total_em_df_long.rename(columns={'country_code_a3': 'iso_alpha_3'}, inplace=True)

total_em_df_long
```

```python
# Remove prefix to year field
total_em_df_long["year"] = total_em_df_long["year"].str.replace("Y_", "", regex=False).astype(int)

# Convert total emissions from Gg to Mt
total_em_df_long["total_emissions"] = total_em_df_long["total_emissions"] / 1000.0
total_em_df_long
```

```python
print("Unique iso_alpha_3 count:", total_em_df_long["iso_alpha_3"].nunique())
print("Year range:", total_em_df_long["year"].min(), "to", total_em_df_long["year"].max())
```

```python
total_em_df_long.to_csv(os.path.join(PROCESSED_DATA_DIR_PATH, 'historical_total_emission_per_country.csv'), index=False)
```
