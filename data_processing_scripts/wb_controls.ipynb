### 📊 Control Variables Overview

### The Goal of the Regression
We are testing whether **climate policy restrictiveness reduces GHG emissions**, ensuring that **differences between countries do not confound the causal relationship**.

### Purpose of Control Variables
Control variables are necessary when they:
1. **Affect both the independent (policy restrictiveness) and dependent (GHG emissions) variables.**
2. **Could bias the estimated effect of policy restrictiveness on emissions.**
3. **Help ensure comparability across countries.**

### 📌 Recommended Control Variables
To account for country-specific **structural, economic, social, and environmental differences**, the following variables are included:

- **Economic**: GDP per capita, industry size, electricity consumption, trade openness, inflation.
- **Energy**: Fossil fuel use, coal consumption, renewable energy.
- **Governance**: Political stability, government effectiveness, corruption.
- **Social**: Poverty, inequality, education, unemployment.
- **Environmental**: Climate vulnerability, air pollution.

### Reviewing Missing Data for All Indicators

```python
### Fetching Indicators & Handling Errors
import wbdata
import pandas as pd

# 🎯 Define control variables (indicators)
indicators = {
    "NY.GDP.PCAP.CD": "GDP per capita",
    "NE.TRD.GNFS.ZS": "Trade openness (% of GDP)",
    "BX.KLT.DINV.WD.GD.ZS": "FDI net inflows (% of GDP)",
    "FP.CPI.TOTL.ZG": "Inflation rate",
    "GC.REV.XGRT.GD.ZS": "Government revenue (% of GDP)",
    "DT.DOD.DECT.GN.ZS": "Foreign debt (% of GDP)",
    "NV.IND.TOTL.ZS": "Industry size (% of GDP)",
    "EG.USE.ELEC.KH.PC": "Electricity consumption per capita",
    "SI.POV.LMIC": "Poverty headcount (% below $3.65/day)",
    "SL.UEM.TOTL.ZS": "Unemployment rate",
    "SI.POV.GINI": "Income inequality (Gini Index)",
    "EN.POP.DNST": "Population density (people per sq km)",
    "SE.SEC.ENRL": "Education level (Secondary enrollment)",
    "VC.IHR.PSRC.P5": "Intentional homicides (per 100,000 people)",
    "PV.EST": "Political stability and absence of violence/terrorism estimate",
    "CC.EST": "Control of corruption estimate",
    "EN.ATM.CO2E.KD.GD": "CO₂ emissions per unit of GDP",
    "EG.USE.PCAP.KG.OE": "Energy use per capita",
    "EG.USE.COMM.FO.ZS": "Fossil fuel dependency (% total energy)",
}

# 🎯 Define time range
start_date = "1990"
end_date = "2024"

# 🎯 Create dictionaries for storing results
successful_data = {}
failed_indicators = []

# 🎯 Fetch each indicator and handle missing data
for key, value in indicators.items():
    try:
        print(f"Fetching {value} ({key})...")
        data = wbdata.get_dataframe({key: value}, country="all", date=(start_date, end_date), freq="Y")

        if data.empty:
            print(f"❌ No data found for {value} ({key})")
            failed_indicators.append((key, value))
        else:
            data.reset_index(inplace=True)
            data["date"] = pd.to_datetime(data["date"]).dt.year  # Ensure year format
            data.rename(columns={"country": "Country"}, inplace=True)  # Ensure country column is available
            successful_data[key] = data  # Store data

    except Exception as e:
        print(f"❌ Error fetching {value} ({key}): {e}")
        failed_indicators.append((key, value))

# 🎯 Save failed indicators summary
failed_summary = pd.DataFrame(failed_indicators, columns=["Indicator Code", "Indicator Name"])
failed_summary.to_csv("failed_indicators_summary.csv", index=False)

# 🎯 Display missing indicator summary
print("\n❌ **Indicators with No Available Data**")
if failed_summary.empty:
    print("✅ All indicators retrieved successfully!")
else:
    from IPython.display import display
    display(failed_summary)
```

```python
# Combining Data & Analyzing Missing Data by Country

# 🎯 Combine all fetched data into a single DataFrame
if successful_data:
    combined_data = pd.concat(successful_data.values(), axis=1)
    combined_data = combined_data.loc[:, ~combined_data.columns.duplicated()]  # Remove duplicate country columns
else:
    combined_data = pd.DataFrame()

# 🎯 Ensure "Country" column is present
if "Country" not in combined_data.columns:
    combined_data.reset_index(inplace=True)
    if "Country" not in combined_data.columns:
        print("ERROR: 'Country' column is missing! Ensure the dataset includes country-level data.")

# 🎯 Missing data analysis **by country**
if "Country" in combined_data.columns:
    missing_data_per_country = combined_data.set_index("Country").isnull().sum(axis=1)
    available_data_per_country = combined_data.set_index("Country").notnull().sum(axis=1)
    total_entries = missing_data_per_country + available_data_per_country
    missing_percentage = (missing_data_per_country / total_entries) * 100

    # 🎯 Create a summary DataFrame
    missing_data_summary = pd.DataFrame({
        "Missing Data Points": missing_data_per_country,
        "Available Data Points": available_data_per_country,
        "% Missing": missing_percentage
    }).reset_index()

    # 🎯 Save missing data summary
    missing_data_summary.to_csv("missing_data_summary_by_country.csv", index=False)

    # 🎯 Display missing data summary
    from IPython.display import display
    print("\n📌 **Missing Data Summary by Country & Indicator**")
    display(missing_data_summary.head(20))
```

```python
# Analyzing Completeness of Indicators Across Countries

import wbdata
import pandas as pd

# 🎯 Define control variables (indicators)
indicators = {
    "NY.GDP.PCAP.CD": "GDP per capita",
    "NE.TRD.GNFS.ZS": "Trade openness (% of GDP)",
    "BX.KLT.DINV.WD.GD.ZS": "FDI net inflows (% of GDP)",
    "FP.CPI.TOTL.ZG": "Inflation rate",
    "GC.REV.XGRT.GD.ZS": "Government revenue (% of GDP)",
    "DT.DOD.DECT.GN.ZS": "Foreign debt (% of GDP)",
    "NV.IND.TOTL.ZS": "Industry size (% of GDP)",
    "EG.USE.ELEC.KH.PC": "Electricity consumption per capita",
    "SI.POV.LMIC": "Poverty headcount (% below $3.65/day)",
    "SL.UEM.TOTL.ZS": "Unemployment rate",
    "SI.POV.GINI": "Income inequality (Gini Index)",
    "EN.POP.DNST": "Population density (people per sq km)",
    "SE.SEC.ENRL": "Education level (Secondary enrollment)",
    "VC.IHR.PSRC.P5": "Intentional homicides (per 100,000 people)",
    "PV.EST": "Political stability and absence of violence/terrorism estimate",
    "CC.EST": "Control of corruption estimate",
    "EN.ATM.CO2E.KD.GD": "CO₂ emissions per unit of GDP",
    "EG.USE.PCAP.KG.OE": "Energy use per capita",
    "EG.USE.COMM.FO.ZS": "Fossil fuel dependency (% total energy)",
}

# 🎯 Define time range
start_date = "1990"
end_date = "2024"

# 🎯 Create lists for storing summaries
summary_list = []

# 🎯 Fetch data and analyze completeness
for key, value in indicators.items():
    try:
        data = wbdata.get_dataframe({key: value}, country="all", date=(start_date, end_date), freq="Y")

        if data.empty:
            raise ValueError(f"No data found for {value} ({key})")  # Raise an error instead of setting 0s

        # 🎯 Reset index for easier handling
        data.reset_index(inplace=True)
        data["date"] = pd.to_datetime(data["date"]).dt.year
        data.rename(columns={"country": "Country"}, inplace=True)

        # Expected years (1990-2024) = 35 years
        expected_years = set(range(1990, 2024))

        # Count available years per country
        available_data = data.dropna().groupby("Country")["date"].apply(set).reset_index()
        available_data["Available Years"] = available_data["date"].apply(len)
        available_data["% Available"] = (available_data["Available Years"] / len(expected_years)) * 100

        # Categorize countries by data availability
        total_countries = len(available_data)
        high_availability = len(available_data[available_data["% Available"] >= 90])
        medium_availability = len(available_data[(available_data["% Available"] >= 60) & (available_data["% Available"] < 90)])
        low_availability = len(available_data[(available_data["% Available"] > 0) & (available_data["% Available"] < 60)])
        no_data = data.groupby("Country")[value].apply(lambda x: x.isnull().all()).reset_index()
        no_data = len(no_data[no_data[value]])

        # 🎯 Append results to summary list
        summary_list.append([value, total_countries, high_availability, medium_availability, low_availability, no_data])

    except Exception as e:
        print(f" Error processing {value} ({key}): {e}")
        summary_list.append([value, "Error", "Error", "Error", "Error", "Error"])  # Mark error cases

# 🎯 Create DataFrame for summary
summary_df = pd.DataFrame(summary_list, columns=["Indicator", "Total Countries", "≥ 90% Available", "60%–90% Available", "≤ 60% Available", "0% Available"])

# 🎯 Save summary to CSV
summary_df.to_csv("indicator_completeness_summary.csv", index=False)

# 🎯 Display summary in Jupyter
from IPython.display import display

print("\n📊 **Indicator Completeness Summary**")
display(summary_df)
```
